```python
s = audiolab.sndfile("11k16bitpcm.wav")
```

```python
frames = s.read_frames(s.get_nframes())
```

```python
audiolab.play(frames, 11000)
```

```python
def show_and_play(start, end):
    mfcc = talkbox.features.mfcc(frames)
    audiolab.play(frames[:1000])
    fig = plt.figure()
    fig.set_size_inches(20, 20)
    ax = fig.add_subplot(111)
    ax.imshow(mfcc[0].transpose()[:, start:end])
show_and_play(100, 200)
```
